# KMeans v6
v6 uses the dataframe without key, time signature, popularity, explicit, duration_ms, speechiness, loudness, liveness

```python
import functions as f
from IPython.display import IFrame
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import MinMaxScaler
```

## Import data

```python
df = pd.read_csv("data/clean_audio_features_dataset_curated.csv")
```

```python
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
df_model = df.drop(columns=['track_id', 'artists', 'album_name', 'track_name']).copy()
```

```python
df_model.columns
```

```python
df_model = df_model[['danceability', 
    'energy',
    'instrumentalness',
    'valence',
    'tempo',
    'acousticness']]
df_model
```

## Scaler

```python
minmax = MinMaxScaler()
```

```python
minmax.fit(df_model)
X_scaled = minmax.transform(df_model)
```

```python
X_scaled
```

## KMeans model metrics

```python
inertias = []
k_range = range(2, 15)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=5)
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)


plt.figure(figsize=(8, 4))
plt.plot(k_range, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (Within-cluster sum of squares)')
plt.title('Elbow Method for Optimal k')
plt.xticks(k_range)
plt.grid(True)
plt.show()
```

```python
cluster_range = range(2, 15)
silhouette_scores = []
for n_clusters in cluster_range:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=5)
    cluster_labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, cluster_labels)
    silhouette_scores.append(score)
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='-', color='b')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Different Cluster Counts")
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

## Model training n_clusters = 5

```python
kmeans = KMeans(n_clusters=5, random_state=42, n_init=5)
```

```python
kmeans.fit(X_scaled)
```

### Model prediction

```python
df_kmeans_v6_5 = pd.read_csv("data/clean_audio_features_dataset_curated.csv")
df_kmeans_v6_5 = df_kmeans_v6_5.drop(columns=['Unnamed: 0'])
df_kmeans_v6_5['cluster'] = kmeans.predict(X_scaled)
```

```python
df_kmeans_v6_5['cluster'].value_counts()
```

### Check clusters

```python
df_kmeans_v6_5[['danceability', 
    'energy',
    'instrumentalness',
    'valence',
    'tempo',
    'acousticness',
    'cluster']].select_dtypes('number').groupby('cluster').mean()
```

```python
def play_song(track_id):
    return IFrame(src="https://open.spotify.com/embed/track/"+track_id,
       width="320",
       height="80",
       frameborder="0",
       allowtransparency="true",
       allow="encrypted-media",
      )

def play_random_from_cluster(dataframe, cluster_number):
    track_id = dataframe[dataframe['cluster']==cluster_number].sample()['track_id'].values[0]
    return play_song(track_id)

def play_random_5percluster(dataframe):
    for i in sorted(dataframe['cluster'].unique()):
        print(f"Songs from cluster {i}:")
        for j in range(5):
            track_id = dataframe[dataframe['cluster']==i].sample()['track_id'].values[0]
            display(play_song(track_id))
```

```python
play_random_from_cluster(df_kmeans_v6_5, 4)
```

```python
play_random_5percluster(df_kmeans_v6_5)
```

Clusters:
- 0: Instrumental
- 1: Electronic sounds
- 2: Sing along
- 3: Slow beats
- 4: Party Mix
